```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Загрузка данных
data = pd.read_csv('Invistico_Airline.csv')
df.info()

# Проверка названий столбцов
print("Названия столбцов в наборе данных:")
print(data.columns)

# имя целевой переменной
target_variable_name = 'satisfaction'  # целевая переменная

# Преобразование целевой переменной в бинарный формат
data[target_variable_name] = np.where(data[target_variable_name] == 'satisfied', 1, 0)

# 1. Построение графика распределения целевой переменной
plt.figure(figsize=(10, 6))
sns.countplot(x=data[target_variable_name])
plt.title('Количество данных по каждому классу целевой переменной')
plt.xlabel('Класс (0 - dissatisfied, 1 - satisfied)')
plt.ylabel('Количество')
plt.savefig('class_distribution.png')
plt.show()

# 2. Построение ящика с усами для числовых признаков
numeric_features = data.select_dtypes(include=[np.number])
plt.figure(figsize=(15, 8))
sns.boxplot(data=numeric_features)
plt.title('Ящик с усами для числовых признаков')
plt.xticks(rotation=45)
plt.savefig('boxplot_numeric_features.png')
plt.show()

# Выявление выбросов
outliers = {}
for col in numeric_features.columns:
    Q1 = numeric_features[col].quantile(0.25)
    Q3 = numeric_features[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers[col] = numeric_features[(numeric_features[col] < lower_bound) | (numeric_features[col] > upper_bound)].shape[0]

print("Число выбросов в числовых признаках:")
print(outliers)

# 3. Анализ текстовых признаков
categorical_features = data.select_dtypes(include=[object])
for col in categorical_features.columns:
    plt.figure(figsize=(10, 6))
    sns.countplot(y=categorical_features[col])
    plt.title(f'Распределение значений для {col}')
    plt.xlabel('Количество')
    plt.ylabel(col)
    plt.savefig(f'distribution_{col}.png')
    plt.show()

    # Проверка на дисбаланс
    value_counts = categorical_features[col].value_counts()
    if value_counts.min() < 0.1 * value_counts.sum():  # Если минимальное количество меньше 10% от общего
        print(f'Признак {col} имеет дисбаланс значений:\n{value_counts}')

# 4. Дополнительный анализ для зафиксированных признаков с дисбалансом
for col in categorical_features.columns:
    value_counts = categorical_features[col].value_counts()
    if value_counts.min() < 0.1 * value_counts.sum():
        plt.figure(figsize=(10, 6))
        sns.countplot(x=categorical_features[col])
        plt.title(f'График количества значений для {col}')
        plt.xlabel(col)
        plt.ylabel('Количество')
        plt.savefig(f'countplot_{col}.png')
        plt.show()
```
